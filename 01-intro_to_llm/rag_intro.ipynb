```python
import minsearch
```

```python
import json
```

```python
with open('documents.json','rt') as f_in:
    docs_raw=json.load(f_in)
```

```python
documents=[]

for course_dict in docs_raw:
    for doc in course_dict['documents']:
        documents.append(doc)
```

```python
index=minsearch.Index( text_fields=["question", "text", "section"],
    keyword_fields=["course"])
```

```python
def search_and_ask():
    print("--- Course FAQ Assistant (type 'exit' to stop) ---")
    
    while True:
        query = input("\nYour Question: ")
        
        if query.lower() == 'exit':
            print("Goodbye!")
            break
            
        # Defining the search parameters based on your setup
        boost = {'question': 3.0, 'section': 0.5}
        
        # Executing the search
        results = index.search(
            query=query,
            boost_dict=boost,
            num_results=3
        )
        
        # Displaying the results
        if not results:
            print("I couldn't find an answer to that.")
        else:
            print(f"\nFound {len(results)} relevant answer(s):")
            for i, res in enumerate(results, 1):
                print(f"\n[{i}] Question: {res['question']}")
                print(f"Answer: {res['text']}")
                print("-" * 20)
```
